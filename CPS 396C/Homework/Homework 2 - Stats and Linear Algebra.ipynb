Consider the [iris dataset](https://archive.ics.uci.edu/ml/datasets/iris).  It is a curated dataset and can be imported with pandas.

```python
import pandas as pd

iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

#print iris
#print type(iris)
print iris.values
print type(iris.values)
```

Let's try to seperate the data by type of iris.  Then create a dictionary with the seperate sets.  

```python
import numpy as np

setosa = np.matrix([r[:-1] for r in iris.values if r[-1] == 'setosa'], dtype=np.float32)
versicolor = np.matrix([r[:-1] for r in iris.values if r[-1] == 'versicolor'], dtype=np.float32)
virginica = np.matrix([r[:-1] for r in iris.values if r[-1] == 'virginica'], dtype=np.float32)

# print type(setosa), setosa.shape

iris = {'setosa': setosa, 'versicolor': versicolor, 'virginica': virginica}
print iris
```

Calculate the mean and standard deviation of each attribute.  

1) Do you see any attributes that might serve to discriminate between the types?

```python
np.set_printoptions(precision=4) # Sets the number of significant digits to display to 4

# print setosa
# print np.mean(setosa, axis=0)
# print np.std(setosa, axis=0)

# Or more concisely 
for typ, data in iris.iteritems():
    print np.mean(data, axis=0)
```

2) Calculate the correlation between each pair of attributes.

```python
for name, data in iris.iteritems():
    print name + '\n'
    completed = np.zeros((data[:][0].size, data[:][0].size))
    
    for i in range(data[:][0].size):
        for j in range(data[:][0].size):
            if i != j and completed[i][j] == 0:
                completed[i][j], completed[j][i] = 1, 1
                
                x = np.transpose(data[:,i])
                y = np.transpose(data[:,j])
                print "Attributes {} & {}: \n{:.2f}".format(i + 1, j + 1, np.corrcoef(x,y)[0][1])
    
    print '\n'
```

3) Have a look at the following data over a particular game.  The first two columns can the input from players 1 and 2 and the third column indicates the winner.  Split off the third column and put it into a `target` vector.  Keep the first two columns and save as `input data`. 

```python
from numpy import genfromtxt
game_data = genfromtxt('http://mlid.cps.cmich.edu/resources/game-data.txt', delimiter=' ')

print game_data

input_data = np.matrix([inp[:-1] for inp in game_data], dtype=np.float32)
target = np.matrix([targ[-1] for targ in game_data], dtype=np.float32)

print input_data.shape, target.shape
print target
```

4) Looking at the data, can you determine a general rule to pick the winner?  What is it?  Use it to predict the winner.

```python
winners = []

print input_data
for combo in input_data:
    x = combo.item(0)
    y = combo.item(1)
    
    if x > y:
        winners.append(0.)
    else:
        winners.append(1.)
        
winners = np.matrix(winners)
print winners
```

Bonus) What percent of the time is your prediction correct (i.e., what is the accuracy)?     

```python
correct_count = 0
total_count = 0

for i in range(winners.size):
    total_count += 1
    
    if winners.item(i) == target.item(i):
        correct_count += 1
        
print "{}% correct".format((correct_count / total_count) * 100)
```
