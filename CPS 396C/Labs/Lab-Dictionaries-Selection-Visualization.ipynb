2) Consider the interests of employees. First, create a bar graph of the frequency of these interests.  Each tuple is a user-interest pair.

```python
interests = [
    (0, "Hadoop"), (0, "Big Data"), (0, "HBase"), (0, "Java"),
    (0, "Spark"), (0, "Storm"), (0, "Cassandra"),
    (1, "NoSQL"), (1, "MongoDB"), (1, "Cassandra"), (1, "HBase"),
    (1, "Postgres"), (2, "Python"), (2, "scikit-learn"), (2, "scipy"),
    (2, "numpy"), (2, "statsmodels"), (2, "pandas"), (3, "R"), (3, "Python"),
    (3, "statistics"), (3, "regression"), (3, "probability"),
    (4, "machine learning"), (4, "regression"), (4, "decision trees"),
    (4, "libsvm"), (5, "Python"), (5, "R"), (5, "Java"), (5, "C++"),
    (5, "Haskell"), (5, "programming languages"), (6, "statistics"),
    (6, "probability"), (6, "mathematics"), (6, "theory"),
    (7, "machine learning"), (7, "scikit-learn"), (7, "Mahout"),
    (7, "neural networks"), (8, "neural networks"), (8, "deep learning"),
    (8, "Big Data"), (8, "artificial intelligence"), (9, "Hadoop"),
    (9, "Java"), (9, "MapReduce"), (9, "Big Data")
]

from collections import Counter

interest_counts = Counter([interest for _, interest in interests])

%matplotlib inline
import matplotlib.pyplot as plt

xs = [3.5 * i for i, _ in enumerate(interest_counts.values())]

plt.bar(xs, interest_counts.values(), 2)
plt.xticks(xs, interest_counts.keys(), rotation='vertical')
plt.ylabel("# of Users")
plt.title("Frequency of Interests")

plt.show()
```

3) Look at some data on company employees as it relates to their time at the company.  Create a scatter plot of salaries and tenure with salary as the dependent variable.  Also, calculate and display the mean salary for 4 ranges of tenure (i.e, [0-3), [3-5), [5-8) and more than 8 years).  

```python
salaries_and_tenures = [(83000, 8.7), (88000, 8.1),
                        (48000, 0.7), (76000, 6),
                        (69000, 6.5), (76000, 7.5),
                        (60000, 2.5), (83000, 10),
                        (48000, 1.9), (63000, 4.2)]

# Create parallel arrays of salaries and tenures
salary, tenure = zip(*salaries_and_tenures)

plt.scatter(tenure, salary)

plt.ylabel("Salary")
plt.xlabel("Tenure")

plt.show()
```
