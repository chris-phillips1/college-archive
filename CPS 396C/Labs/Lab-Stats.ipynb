Consider the [iris dataset](https://archive.ics.uci.edu/ml/datasets/iris).  It is a curated dataset and can be imported with pandas.

Let's try to seperate the data by type of iris.

```python
import numpy as np

# for r in iris.values:
#    if r[-1] == 'setosa':
#        print r 

setosa = np.matrix([r[:-1] for r in iris.values if r[-1] == 'setosa'], dtype = np.float32)
versicolor = np.matrix([r[:-1] for r in iris.values if r[-1] == 'versicolor'], dtype = np.float32)
virginica = np.matrix([r[:-1] for r in iris.values if r[-1] == 'virginica'], dtype = np.float32)
iris = {'setosa': setosa,
        'versicolor': versicolor,
        'virginica': virginica}
print iris
```
