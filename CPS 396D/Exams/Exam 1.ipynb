Your task in this exam is to create a line plot for the following equation over the interval [0, 10$\pi$], inclusive:

\begin{equation*}
f(x) = 3 * sin(x) + 2 * cos(x/2)
\end{equation*}

### Problem 1 (10 Points)

Write out the magic function for displaying plots inline in this notebook, and put all of your package import statements here. You will be creating numpy arrays, doing mathematical operations and plotting in this exam.

```python
%matplotlib inline

import numpy as np
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Problem 2 (30 Points)

Create 2 numpy arrays for $x$ and $f(x)$. Both arrays should be 200 elements in length. The $x$ array should contain values equally spaces between $0$ and $10\pi$ (10 points). The $f(x)$ or $y$ array should contain the appropriate values from the equation above for each $x$ (20 points).

```python
numElems = 200
xLocs = np.linspace(0., 10. * math.pi, numElems)
yLocs = 3 * np.sin(xLocs) + 2. * np.cos(xLocs / 2)
```

### Problem 3 (40 Points)
Draw a plot of the equation about displaying the specified $x$ range. Add a title, x and y axis labels, and major grid lines spaced every 1 unit in the y axis and 2 units in the x axis. Also, make minor grid lines spaced 0.5 units in each direction.

```python
fig, ax = plt.subplots()
ax.plot(xLocs, yLocs)

ax.set_title("Exam 1")
ax.set_xlabel("X")
ax.set_ylabel("Y")
x_major_ticker = mpl.ticker.MultipleLocator(2.0)
x_minor_ticker = mpl.ticker.MultipleLocator(0.5)
y_major_ticker = mpl.ticker.MultipleLocator(1.0)
y_minor_ticker = mpl.ticker.MultipleLocator(0.5)
ax.xaxis.set_major_locator(x_major_ticker)
ax.yaxis.set_major_locator(y_major_ticker)
ax.xaxis.set_minor_locator(x_minor_ticker)
ax.yaxis.set_minor_locator(y_minor_ticker)

ax.grid(color="black", which="major", axis='x', linestyle='-', linewidth=0.25)
ax.grid(color="black", which="major", axis='y', linestyle='-', linewidth=0.25)
```

### Problem 4 (10 Points)
Based on the plot or on the array values themselves, what is the maximum $f(x)$ value for this equation? Your answer should display no more than 4 significant digits after the decimal.

```python
eqMax = max(yLocs)
print("{:.4f}".format(eqMax))
```
