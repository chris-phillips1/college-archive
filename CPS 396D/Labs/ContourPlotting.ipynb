Look at the following equation for this lab:
    
\begin{equation*}
F(x, y) = 2(x - 1) * 5(y + 1)^2
\end{equation*}


## 1. Contour Plot

Plot a contour plot of this function over the range $x = [-10, 10]$ and $y = [-10, 10]$

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(6, 5))

x = y = range(-10,11)
X, Y = np.meshgrid(x, y)

Z = (2 * (X-1)) * (5 * (Y + 1)**2)

c = ax.contour(X, Y, Z, 50, cmap=mpl.cm.RdBu)
```

## 2. 3D Plot

Plot a 3D plot of this function over the range $x = [-10, 10]$ and $y = [-10, 10]$

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig, axes = plt.subplots(subplot_kw={'projection': '3d'})

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = axes.plot_surface(X, Y, Z, rstride=1, cstride=1, linewidth=0, antialiased=False, norm=norm, cmap=mpl.cm.Blues)
cb = fig.colorbar(p, ax=axes, shrink=0.6)
axes.set_title("Surface Plot of Equation")
axes.set_xlabel("$x$", fontsize=16)
axes.set_ylabel("$y$", fontsize=16)
axes.set_zlabel("$z$", fontsize=16)
```

## 3. Filled Contour Plot

Plot a contour plot of this function over the range $x = [-10, 10]$ and $y = [-10, 10]$. Add axis labelx, a title and a color legend.

```python
fig, ax = plt.subplots(figsize=(6, 5))

x = y = range(-10,11)
X, Y = np.meshgrid(x, y)

Z = (2 * (X-1)) * (5 * (Y + 1)**2)

c = ax.contourf(X, Y, Z, 50, cmap=mpl.cm.RdBu)
fig.colorbar(c)

axes.set_title("Contour Plot")
axes.set_xlabel("$x$", fontsize = 16)
```
