```python
import random

class Door:
    
    def __init__(self, hidden):
        self.hidden = hidden
    
    def __repr__(self):
        return self.hidden
    
    
class Game:
    
    def __init__(self):
        self.car_door = random.randint(0,2)
        self.shown_door = -1
        self.user_door = -1
        self.move_door = -1
        
        self.turn = 0
        self.move = ''
        self.win = False
    
    def initalize(self):
        doors = []
        
        for i in range(3):
            if i is self.car_door:
                doors.append(Door('car'))
            else:
                doors.append(Door('goat'))
                      
    def play_game(self):
        
        if self.turn is 0:
            print('Choose a door!')
            print('|| 1 || 2 || 3 ||')
            self.user_door = int(input()) - 1
            self.turn = 1
            self.play_game()
        elif self.turn is 1:
            self.show_door()
            print('Door {} has a goat behind it.'.format(str(self.shown_door + 1)))
            self.move = input('Stay or Move? (S or M): ')
            self.evaluate_game()
            self.turn = 2
        else:
            return
        
    def calculate_move_door(self):
        if self.user_door is 0:
            if self.shown_door is 1:
                self.move_door = 2
            else:
                self.move_door = 1
        elif self.user_door is 1:
            if self.shown_door is 0:
                self.move_door = 2
            else:
                self.move_door = 0
        elif self.user_door is 2:
            if self.shown_door is 0:
                self.move_door = 1
            else:
                self.move_door = 0
        
    def evaluate_game(self):
        if self.move is 'M':
            self.calculate_move_door()
            if self.move_door is self.car_door:
                self.win = True
                print('You won! The car is behind door {}\n'.format(self.car_door + 1))   
            else:
                print('You lost, your move cost you the game\n')
        elif self.move is 'S':
            if self.user_door is self.car_door:
                self.win = True
                print('You won! The car is behind door {}\n'.format(self.car_door + 1))
            else:
                print('You lost, the car was behind door {}\n'.format(self.car_door + 1))
            
    def show_door(self):
        if self.user_door is 0:
            self.shown_door = 1 if 1 is not self.car_door else 2
        elif self.user_door is 1:
            self.shown_door = 0 if 0 is not self.car_door else 2
        elif self.user_door is 2:
            self.shown_door = 0 if 0 is not self.car_door else 1

            
for i in range(5):
    g = Game()
    g.initalize()
    g.play_game()
                
```
